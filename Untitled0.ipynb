```python
df['GE'] = 0
df.loc[df['Gender']=='Female','GE'] = 1
df.loc[df['Gender']=='Male','GE'] = 2
df.loc[df['GE']==0]
```

Am preprocesat coloana **Gender**

```python
df.drop(columns=['Gender', 'Loan_ID'], inplace=True)
```

```python
df
```

```python
df['Married'].value_counts()
```

```python
# df['ME'] = 0
# df.loc[df['Married']=='Yes' , 'ME'] = 1
#df
df['Married'] = df['Married'].map({'No' : 0, 'Yes' : 1})
```

```python
df['Dependents'].value_counts()
```

```python
df['Dependents'] = df['Dependents'].map({'0': 0, '1': 1, '2': 2, '3+': 3})
df
```

```python
df['Education'] = df['Education'].map({'Graduate': 1, 'Not Graduate': 0})
df
```

```python
df['Property_Area'].value_counts()
```

```python
df=pd.get_dummies(df, columns=['Property_Area'])
```

Transformam Property_Area cu 3 valori diferite in 3 coloane bool deoarece daca am pune 1/2/3 ar putea crea probleme in calcul

```python
df.info()
```

```python
df.loc[df['LoanAmount'].isna(), 'LoanAmount'] = df['LoanAmount'].mean() #Unde avem valoare necunoscuta (NaN <=> .isna() == True) punem valoarea medie de pe col (in loc sa stergem)
df.loc[df['Loan_Amount_Term'].isna(), 'Loan_Amount_Term'] = df['Loan_Amount_Term'].mean()
df.loc[df['Married'].isna(), 'Married'] = df['Married'].mean()
df.loc[df['Credit_History'].isna(), 'Credit_History'] = df['Credit_History'].mean()
df.info()

```

```python
from sklearn.neighbors import KNeighborsClassifier
df
```

```python
X = df[ [
    'Married',
    'Dependents',
    'Education',
    'ApplicantIncome',
    'CoapplicantIncome',
    'LoanAmount',
    'Loan_Amount_Term',
    'Credit_History',
    'Property_Area_Rural',
    'Property_Area_Semiurban',
    'Property_Area_Urban'
] ]
X
```

```python
y = df['Loan_Status'].map({'Y': 1,'N': 0})
```

```python
X.loc[df['Dependents'].isna(), 'Dependents'] = X['Dependents'].mode()[0] #datele necunoscute le atribuim ca fiind cea mai des aparuta valoare
X.info()
```

```python
model = KNeighborsClassifier()
model.fit(X,y)
```

```python
model.score(X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X, y)
model.score(X, y)
```
