```python
from sklearn.feature_extraction.text import CountVectorizer

class_names=['BOOKS', 'CLOTHING']
X_train = ['i love the book', 'this is a great book', 'the fit is great', 'i love the shoes']
y_train = [0, 0, 1, 1]
vectorizer = CountVectorizer(ngram_range=(1,2))

X_train_vectors = vectorizer.fit_transform(X_train)
print(vectorizer.get_feature_names_out())
for i in range(4):
  print(X_train_vectors[i].toarray())
```

```python
from sklearn import svm

model = svm.SVC(kernel='linear')
model.fit(X_train_vectors, y_train)
```

```python
X_test = vectorizer.transform(['I like the book', 'Shoes are cool', 'Where did you get that book','The love for the book', 'Fit the book on the shelf', 'Nice fit bro', 'I love the fit of that book in my life', 'I love the books'])
y_test = model.predict(X_test)
for pred in y_test:
  print(class_names[pred])
```

```python
import spacy
#!python -m spacy download en_core_web_md
```

```python
nlp = spacy.load('en_core_web_md')
```

```python
docs = []
X_train = ['i love the book', 'this is a great book', 'the fit is great', 'i love the shoes']
y_train = [0, 0, 1, 1]
for text in X_train:
  docs.append(nlp(text))
X_train_word_vectors = [x.vector for x in docs]
```

```python
model_1 = svm.SVC(kernel='linear')

model_1.fit(X_train_word_vectors, y_train)
```

```python
X_word_test = ['I like the book', 'Shoes are cool', 'Where did you get that book','The love for the book', 'Fit the book on the shelf', 'Nice fit bro', 'I love the fit of that book in my life', 'I love the books',
               'I love the story', 'The library is so nice', 'I like that red scarf', 'Those earrings are ugly', 'I want to read that']
test_docs = [nlp(text) for text in X_word_test]
X_test_word_vectors = [x.vector for x in test_docs]

y_pred = model_1.predict(X_test_word_vectors)
for i in range(len(y_pred)):
  print(X_word_test[i], class_names[y_pred[i]])
```

```python
import re #regex

regexp = re.compile('ab[\S]*cd')  # ab; [^\s] = not \s = not whitespace = [\S] ; * = at least 0 times; cd => starts with ab, at least 0 chars that are not whitespace(' '); cd

phrases = ['abcd', 'xxx', 'abdofiofcd', 'ab cd', 'the silly abbccd amogus']
matches = []
for phrase in phrases:
  if re.match(regexp, phrase):
    matches.append(phrase)
print(matches)
matches2 = []
for phrase in phrases:
  if re.search(regexp, phrase):
    matches2.append(phrase)
print(matches2)
```

```python
regexp = re.compile(r'hat|book|cook')
regexp2 = re.compile(r'\bhat\b|\bbook\b|\bcook\b') #\b = word boundary (requires that its a word of itself)
phrases = ['The rat stole my hat.', 'Nice book mate','Cook for me','I booked the apartment', 'What is wrong with you']
matches = []
matches2 = []
for phrase in phrases:
  if re.search(regexp, phrase):
    matches.append(phrase)
  if re.search(regexp2, phrase):
    matches2.append(phrase)
print(matches)
print(matches2)
```

```python
import nltk

nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

phrase = "I'm really interested in those books of yours. From where have you bought them?"

words = word_tokenize(phrase)
print(f"Words: {words}")
stemmed_words = []
for word in words:
  stemmed_words.append(stemmer.stem(word))
print(f"Stemmed words: {stemmed_words}")
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_words = []
for word in words:
  lemmatized_words.append(lemmatizer.lemmatize(word))
print(phrase)
print(f"Lemmatized words: {lemmatized_words}")
```

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
print(stop_words)
stripped_phrase = []
for word in words:
  if word not in stop_words:
    stripped_phrase.append(word)
print(stripped_phrase) #I was not removed because uppercase. What if we first stem?
```

```python
stripped_stemmed_phrase = []
for word in stemmed_words:
  if word not in stop_words:
    stripped_stemmed_phrase.append(word)
print(stripped_stemmed_phrase)
```
