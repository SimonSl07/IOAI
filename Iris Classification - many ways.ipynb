```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn import naive_bayes
from sklearn.preprocessing import scale
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
df = load_iris()
X = df.data
y = df.target
```

```python
listmodels = [RandomForestClassifier(n_estimators=10),RandomForestClassifier(n_estimators=30), SVC(kernel = 'poly'), SVC(kernel = 'linear'), naive_bayes.MultinomialNB(), naive_bayes.GaussianNB(),
              KNeighborsClassifier()]
for model in listmodels:
  acc = 0
  k = 20
  for _ in range(k):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = .8)
    model.fit(X_train, y_train)
    acc = acc + model.score(X_test, y_test)
  acc = acc / k
  print(model, "Has a score of:       ", acc)

model = LogisticRegression()
X = scale(X)
acc = 0
k = 20
for _ in range(k):
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = .8)
  model.fit(X_train, y_train)
  acc = acc + model.score(X_test, y_test)
acc = acc / k
print(model, "Has a score of:       ", acc)
```
